```python
import nltk
```

```python
from nltk.tokenize import word_tokenize
f = open('ScrapedData.txt', 'r')
raw_text = f.read()
print(raw_text[:1000])
f.close()
```

```python
tokens = word_tokenize(raw_text)
print(tokens[:50])
tokens = list(set(tokens))
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
for word in tokens:
    if word in stop_words:
        tokens.remove(word)
```

```python
# Length of bag-of-words / vocabulary
print(len(tokens))
```

```python
documents = []
print(os.chdir(os.getcwd() + "/NatGeo"))
files = [f for f in os.listdir('.') if os.path.isfile(f)]
for fi in files:
    fii = open(fi, 'r')
    file_text = fii.read()
    file_tokens = word_tokenize(file_text)
    file_tokens = list(set(file_tokens))
    for word in file_tokens:
        if word in stop_words:
            file_tokens.remove(word)
    documents.append(file_tokens)
    fii.close()
```

```python
print(len(documents))
print(documents[0])
```
