```python
#TASK 1: BASIC TEXT PROCESSING PIPELINE
import nltk
texts = """ Paragraph On Friendship: A friend keeps a lot of importance in our life, and everybody enjoys the company of a friend. True friendship is tough to get. Through every hardship and failure, the faithful friend will stand by. They will care for you every time, and gaining a true friendship is a real present."""

for text in texts:
    sentences = nltk.sent_tokenize(texts)
print(sentences)
```

```python
import nltk
texts = """ Paragraph On Friendship: A friend keeps a lot of importance in our life, and everybody enjoys the company of a friend. True friendship is tough to get. Through every hardship and failure, the faithful friend will stand by. They will care for you every time, and gaining a true friendship is a real present."""

for text in texts:
    sentences = nltk.sent_tokenize(texts)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
        tagged = nltk.pos_tag(words)
        print(tagged)
```
